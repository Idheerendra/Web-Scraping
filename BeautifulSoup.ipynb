This is the original method for data scraping.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from bs4 import BeautifulSoup
import requests
reviews =  []
for j in range(2,200):
    url ="https://magicpin.in/Bhopal/Hoshangabad-Road/Restaurant/Vrindavan-Dhaba/store/20bc99/reviews?page={}".format(j)
    data= requests.get(url)
    soup=BeautifulSoup(data.text,"html.parser")
    for i in soup.find_all("p",class_="review"):
        reviews.append(i.text.strip())
```

Convert it into Dataframe

```python
df=pd.DataFrame(reviews,columns=["Reviews"])
```
